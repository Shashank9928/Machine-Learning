```python
#### For Linear Algebra
import numpy as np
#### For Data Processing/manipulation
import pandas as pd
#### For Data visualization
import seaborn as se
import matplotlib.pyplot as mlt
```

```python
#### Loading The Data
data = pd.read_csv("train.csv")
```

```python
data.select_dtypes('number').corrwith(data['SalePrice']).sort_values(ascending=False)
```

```python
mlt.figure(figsize=(10,6))
se.boxplot(x='OverallQual', y='SalePrice', data=data)
mlt.show()
```

```python
se.distplot(data["SalePrice"])
mlt.show()
```

```python
all_data_na = (data.isnull().sum() / len(data)) * 100
all_data_na = all_data_na.drop(all_data_na[all_data_na == 0].index).sort_values(ascending=False)
missing_data = pd.DataFrame(all_data_na,columns=["Missing Percentage"])
missing_data
```

```python
mlt.figure(figsize=(10,5))
mlt.xticks(rotation=90)
se.barplot(x=all_data_na.index, y=all_data_na)
mlt.show()
```
