### Content
#### Fields in the dataset:

    * Name: Name of cereal
    * mfr: Manufacturer of cereal
    * A = American Home Food Products;
    * G = General Mills
    * K = Kelloggs
    * N = Nabisco
    * P = Post
    * Q = Quaker Oats
    * R = Ralston Purina


#### type:
    * cold
    * hot
* calories: calories per serving
* protein: grams of protein
* fat: grams of fat
* sodium: milligrams of sodium
* fiber: grams of dietary fiber
* carbo: grams of complex carbohydrates
* sugars: grams of sugars
* potass: milligrams of potassium
* vitamins: vitamins and minerals - 0, 25, or 100, indicating the typical percentage of FDA recommended
* shelf: display shelf (1, 2, or 3, counting from the floor)
* weight: weight in ounces of one serving
* cups: number of cups in one serving
* rating: a rating of the cereals (Possibly from Consumer Reports?)

```python
data.replace('N','Nabisco',inplace=True)
data.replace('Q','Quaker Oats',inplace=True)
data.replace('K','Kelloggs',inplace=True)
data.replace('R','Raslston Purina',inplace=True)
data.replace('G','General Mills',inplace=True)
data.replace('P','Post',inplace=True)
data.replace('A','American Home Foods Products',inplace=True)
data
```

```python
se.countplot(y=data['mfr'])
mlt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['mfr'] = le.fit_transform(data['mfr'])
data
```

```python
x=data[["calories","sugars","protein","carbo","fiber","mfr"]]
y=data["rating"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=15)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.intercept_
```

```python
y_predict = lr.predict(x_test)
y_predict
```

```python
mlt.scatter(y_predict,y_test)
mlt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,max_error
print("mean_squared_error: ",mean_squared_error(y_test,y_predict))
print("mean_absolute_error: ",mean_absolute_error(y_test,y_predict))
print("root_mean_squared_error: ",mean_squared_error(y_test,y_predict,squared=False))
```

```python
print("max Error: ",max_error(y_test,y_predict))
print("R2_Score: ",r2_score(y_test,y_predict))
```

```python
residual_test = y_test - y_predict

se.distplot(residual_test)
mlt.show()
```

```python
se.boxplot(residual_test)
mlt.show()
```
