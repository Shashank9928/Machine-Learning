# Boston Housing Dataset :

The Boston housing data by Harrison and Rubinfeld (1979) and explore which factors affect the median value of homes.

#### About the Data variables: 
Features : 
* crim - per capita crime rate by town.
* zn - proportion of residential land zoned for lots over 25,000 sq.ft
* indus - proportion of non-retail business acres per town
* chas - Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
* nox - nitric oxides concentration (parts per 10 million)
* rm - average number of rooms per dwelling
* age - proportion of owner-occupied units built prior to 1940
* dis - weighted distances to five Boston employment centres
* rad - index of accessibility to radial highways
* tax - full-value property-tax rate per USD 10,000
* ptratio - pupil-teacher ratio by town
* black - proportion of blacks by town
* lstat - percentage of lower status of the population

Target (dependent) :
* medv - median value of owner-occupied homes in USD 1000â€™s

## - Heatmap :
* The correlation coefficient ranges from -1 to 1. If the value is close to 1, it means that there is a strong positive correlation between the two variables. When it is close to -1, the variables have a strong negative correlation.

* We will select those features which have a high correlation with our target variable 'medv'.


```python
t1 = df.corr()
plt.figure(figsize=(15,10))
ax = sb.heatmap(t1,annot=True, linewidths=.5,cmap="YlGnBu")
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```

### Observation :
* From the correlation matrix we can see that 'rm' has a strong positive correlation with 'medv' (0.7), where as 'lstat' has a high negative correlation with 'medv' (-0.74).
* The features 'rad', 'tax' pairs are strongly correlated to each other. So, we will not select these both feature.
* And We will not select 'chas' and 'rad' because they are not continous.
* We will not select the features 'crim', 'zn' and 'black' because they have lot of outliers.
* Based on the above observations we will take 'rm', 'lstat' and 'ptratio' as our features.
* And also based on above visualization, we will do the further analysis. 

```python
sb.set(style='darkgrid')
sb.lmplot(x = 'medv',y='rm',data=df)
plt.show()
```

### Observation : 
* The prices increase as the value of 'rm' increases linearly. 
* There are few outliers.

```python
sb.set(style='darkgrid')
sb.lmplot(x = 'medv',y='lstat',data=df)
plt.show()
```

### Observation :

# 4. Data Modeling :

##  - Data Splitting :-
* Now we use the train_test_split function from sklearn library to split our dataset.

```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['rm','lstat','ptratio']]
y = df['medv']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.25,random_state=10)
```

```python
# y_score = lm.score(x_test,y_test)
# print(y_score)
```

* Here we are taking 75% of data for training and 25% data for testing.
* Random State just fix the randomness data state which will not change.

## - Training and testing the model:-
* Now we use scikit-learnâ€™s LinearRegression algorithm to train our model on both the training and test sets.

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train)
```

* Above it fits the model based on x_train and y_train.

## - Prediction :-
* Now we try to predict it to the x_test.
* And the prediction values which will be stored into y_predict.

```python
y_predict = lm.predict(x_test)
```

## - Model evaluation :-
* We will evaluate our model using RMSE and R2-score.

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
r2_score()
```

```python
mae = mean_absolute_error(y_test,y_predict)
rmse = (np.sqrt(mean_squared_error(y_test,y_predict)))
# rmse1 = mean_squared_error(y_test,y_predict)
score = r2_score(y_predict,y_test)
print('Mean absolute error is {}'.format(mae))
# print(rmse1)
print('Root Mean Square Error is {}'.format(rmse))
print('R^2 score is {}'.format(score))
```
